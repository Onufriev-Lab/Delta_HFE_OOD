```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../expt_phys_data/full_dataset_with_scaffold_split.csv').fillna('')
df
```

```python
df['uncertain_expt_value'] = 0
```

```python
for i, row in df.iterrows():
    if(abs(row.expt) < 6):
        if(row.uncertainty > 0.6):
            df.at[i,'uncertain_expt_value'] = 1
    else:
        if((row.uncertainty/abs(row.expt)) > 0.1):
            df.at[i,'uncertain_expt_value'] = 1
```

```python
df[df.uncertain_expt_value == 0].describe()
```

```python
df.to_csv('../expt_phys_data/full_dataset_with_scaffold_split.csv',index=False)
```
